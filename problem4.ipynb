```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

plt.rc('font', family='Times New Roman', size=20)
labelfont = {
    'family' : 'Times New Roman',
    'weight' : 'normal',
    'size'   : 20,
}
```

```python
count = 5
df_C= pd.read_csv("C.csv", names=[str(i) for i in range(1, count + 1)])
df_S = pd.read_csv("S.csv", names=[str(i) for i in range(1, count + 1)])
```

```python
fig, ax = plt.subplots(2, 1, figsize=(11.5,8))
plt.tight_layout(pad=3) # 间距
titles = ["Concentration", "Spectrum"]

ax[0].tick_params(labelsize=18) # tick
for j in range(count):
    ax[0].plot(list(range(1, df_C.shape[0] + 1)), df_C.iloc[:, [j]], linewidth=2.0, label=str(j))
ax[0].set_title(titles[0])
ax[0].legend(loc='upper left',prop=labelfont)

ax[1].tick_params(labelsize=18) # tick
for j in range(count):
    ax[1].plot(list(range(1, df_S.shape[0] + 1)), df_S.iloc[:, [j]], linewidth=2.0, label=str(j))
ax[1].set_title(titles[1])
ax[1].legend(loc='upper left',prop=labelfont)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

plt.rc('font', family='Times New Roman', size=20)
labelfont = {
    'family' : 'Times New Roman',
    'weight' : 'normal',
    'size'   : 20,
}
```

```python
data = pd.read_csv("denoise_result.csv")
```

```python
fig, ax = plt.subplots(3, 1, figsize=(11.5,8))
plt.tight_layout(pad=3) # 
titles = ["Original data", "Denoise data", "Detrend data"]
for i in range(3):
    ax[i].tick_params(labelsize=18) # tick
    ax[i].plot(data.loc[:, ["rt"]], data.iloc[:, [i + 1]], linewidth=2.0, label='signal')
    ax[i].set_title(titles[i])
    ax[i].legend(loc='upper left',prop=labelfont)
```
